```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("heart.csv")

print("ПЕРВЫЙ ВЗГЛЯД НА ДАННЫЕ")
print(f"Размер: {df.shape[0]} строк, {df.shape[1]} столбцов\n")
print("Первые 5 строк:")
print(df.head())
```

```python
print("ТИПЫ ДАННЫХ:")
df.info()

print("\nПРОПУСКИ В ДАННЫХ:")
print(df.isnull().sum())
```

```python
print("ОПИСАТЕЛЬНАЯ СТАТИСТИКА:")
print(df.describe().round(2))
```

```python
print("БАЛАНС КЛАССОВ (target):")
counts = df['target'].value_counts()
print(f"Здоров (0): {counts[0]} ({counts[0]/len(df)*100:.1f}%)")
print(f"Болен  (1): {counts[1]} ({counts[1]/len(df)*100:.1f}%)")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

sns.histplot(data=df, x='age', hue='target', bins=30, kde=True, ax=axes[0,0])
axes[0,0].set_title('Возраст')

sns.histplot(data=df, x='trestbps', hue='target', bins=30, kde=True, ax=axes[0,1])
axes[0,1].set_title('Давление в покое')

sns.histplot(data=df, x='chol', hue='target', bins=30, kde=True, ax=axes[0,2])
axes[0,2].set_title('Холестерин')

sns.histplot(data=df, x='thalach', hue='target', bins=30, kde=True, ax=axes[1,0])
axes[1,0].set_title('Максимальный пульс')

sns.histplot(data=df, x='oldpeak', hue='target', bins=30, kde=True, ax=axes[1,1])
axes[1,1].set_title('Депрессия ST')

sns.histplot(data=df, x='ca', hue='target', bins=8, ax=axes[1,2])
axes[1,2].set_title('Количество сосудов')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.boxplot(data=df, x='target', y='age', ax=axes[0,0])
axes[0,0].set_title('Возраст vs Диагноз')
axes[0,0].set_xticklabels(['Здоров', 'Болен'])

sns.boxplot(data=df, x='target', y='trestbps', ax=axes[0,1])
axes[0,1].set_title('Давление vs Диагноз')
axes[0,1].set_xticklabels(['Здоров', 'Болен'])

sns.boxplot(data=df, x='target', y='chol', ax=axes[1,0])
axes[1,0].set_title('Холестерин vs Диагноз')
axes[1,0].set_xticklabels(['Здоров', 'Болен'])

sns.boxplot(data=df, x='target', y='thalach', ax=axes[1,1])
axes[1,1].set_title('Макс. пульс vs Диагноз')
axes[1,1].set_xticklabels(['Здоров', 'Болен'])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.countplot(data=df, x='sex', hue='target', ax=axes[0,0])
axes[0,0].set_title('Пол')
axes[0,0].set_xticklabels(['Жен', 'Муж'])

sns.countplot(data=df, x='cp', hue='target', ax=axes[0,1])
axes[0,1].set_title('Тип боли в груди')

sns.countplot(data=df, x='exang', hue='target', ax=axes[1,0])
axes[1,0].set_title('Стенокардия')
axes[1,0].set_xticklabels(['Нет', 'Да'])

sns.countplot(data=df, x='fbs', hue='target', ax=axes[1,1])
axes[1,1].set_title('Сахар в крови > 120')
axes[1,1].set_xticklabels(['Нет', 'Да'])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True)
plt.title('Матрица корреляций признаков')
plt.tight_layout()
plt.show()

print("Корреляция признаков с target:")
print(corr['target'].sort_values(ascending=False))
```
